# Metrics 

 The collection .csv is already on the database. This is for defining metrics and validating them subsequently in the Tableau Dashboard.

```python
import pandas as pd
from pathlib import Path

def query_data():
    data = Path('data/processed/merged_collection.csv')
    df = pd.read_csv(data)
    return df

def subset_data() -> pd.DataFrame:
    df = query_data()

    cols = ['name','quantity', 'set', 'type_line', 'cmc', 'mana_cost', 'keywords', 'color_identity', 'colors', 'power', 'toughness', 'loyalty', 'legalities', 'edhrec_rank']

    core_data = [col for col in df.columns if col in cols]

    df = df[core_data]
    return df

def metrics():
    df = subset_data()

    #Number of Unique cards
    print(f"Unique cards: {df['name_x'].nunique()}")

    #Most common set
    print(f"Most common set: {df['set_name_x'].value_counts().idxmax()} with {df['set_name_x'].value_counts().max()} cards")

    #Average CMC
    print(f"Average CMC: {df['cmc'].mean()}")

    #Average EDHREC rank
    print(f"Average EDHREC rank: {df['edhrec_rank'].mean()}")

    #Average quantity per card
    print(f"Average quantity per card: {df['quantity'].mean()}")

    #Most common card type
    print(f"Most common card type: {df['type_line'].value_counts().idxmax()} with {df['type_line'].value_counts().max()} cards")

    #Most common color identity
    print(f"Most common color identity: {df['color_identity'].value_counts().idxmax()} with {df['color_identity'].value_counts().max()} cards")

    #Most common keyword
    print(f"Most common keyword: {df['keywords'].value_counts().idxmax()} with {df['keywords'].value_counts().max()} cards")

    #Most common mana cost
    print(f"Most common mana cost: {df['mana_cost'].value_counts().idxmax()} with {df['mana_cost'].value_counts().max()} cards")

    #Most common card
    print(f"Most common card: {df['name_x'].value_counts().idxmax()} with {df['name_x'].value_counts().max()} copies")

    #Distribution of card types
    print(f"Distribution of card types: {df['type_line'].value_counts(normalize=True)}")

    #Distribution of color identities
    print(f"Distribution of color identities: {df['color_identity'].value_counts(normalize=True)}")

    #Distribution of keywords
    print(f"Distribution of keywords: {df['keywords'].value_counts(normalize=True)}")



metrics()
```
